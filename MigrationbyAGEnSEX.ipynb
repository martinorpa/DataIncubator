```r
reset_par()
library(lattice)
y1 <- migrationage[21:30,]$Age04
y2 <- migrationage[21:30,]$Age59
y3 <- migrationage[21:30,]$Age915
y4 <- migrationage[21:30,]$Age1519
yy <- c(y1,y2,y3,y4)
zz <- migrationage[21:30,]$Area
#plot(x, y1,ylim = range(c(y1,y2,y3,y4)))
plot(zz, cbind (y1, y2,y3,y4), pch = 1, xlab = "Country", ylab = "Number of People(Thousands)",col = "gray",labels=row.names(zz))
points(zz,y1,pch=20,col="Red")
points(zz,y2,pch=25, col="Green")
points(zz,y3,pch=25, col="Lightblue")
points(zz,y4,pch=25, col="Orange")
axis(1, at=seq(1,10,2), labels=letters[1:5])
```
